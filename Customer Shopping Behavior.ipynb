<h3>Analyzing in Python</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Z:\Power BI\Project-1\customer_shopping_behavior.csv")
```

```python
# Here we are find the meadian review rating in each category(like shoes, clothes, etc) to fill the missing or null smaces in reviews this will help us to fill correct review in correct category
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Changing everything into lower case and replacing spaces with _ for to make it in snake casing for better code structure

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
```

```python
df.columns
```

```python
# changing "purchase_amount_(usd)" to just "purchase_amount"

df = df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
# creating a column age_gape to categorise the data based on age groups created by us

labels = ["Young Adult","Adult","Middle-aged","senior"]
df["age_group"]=pd.qcut(df["age"], q=4, labels = labels)
```

```python
df[["age","age_group"]].head(10)
```

```python
# Creating columns Purchase_frequency_days

frequency_mapping = {
"Fortnightly": 14,
"Weekly": 7,
"Monthly": 30,
"Quarterly": 90,
"Bi-Weekly": 14,
"Annually": 365,
"Every 3 Months": 90
}
df ["purchase_frequency_days"] = df ["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchase_frequency_days","frequency_of_purchases"]].head(10)
```

```python
# checking if these columns have same data

df[["discount_applied","promo_code_used"]].head(10)
```

```python
# Checking using the code, output is TRUE that means data is same now we can delete one column

(df["discount_applied"] == df["promo_code_used"]).all()
```

```python
df = df.drop("promo_code_used", axis=1)  # axis=1 drops column (horizontal), axis=0 drops row (vertical)
```

```python
from sqlalchemy import create_engine, text

engine = create_engine(
 "mysql+pymysql://root:rootuser@localhost:3306/pandasdb",
 echo=True # shows SQL logs (optional)
)

# Test connection
with engine.connect() as conn:
 result = conn.execute(text("SELECT DATABASE();"))
 print("Connected to:", result.fetchone())
```
